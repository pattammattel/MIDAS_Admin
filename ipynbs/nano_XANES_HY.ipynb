```python
from sklearn import linear_model
from skimage import io
import numpy as np
from matplotlib import pyplot as plt

dataset = io.imread('C:\\My Work\\Computation Programs\\MatLab Code\\Nanospectroscopy\\Ajith\\Fe_P2.tiff')
y = np.array(dataset)
y = np.reshape(y,[77,120*80])
y[np.isnan(y)] = 0
s1 = np.loadtxt('C:\\My Work\\Computation Programs\\MatLab Code\\Nanospectroscopy\\Ajith\\SteelNP_Bulk_norm_77pts.txt')
s2 = np.loadtxt('C:\\My Work\\Computation Programs\\MatLab Code\\Nanospectroscopy\\Ajith\\hematite_Bulk_norm_77pts.txt')
```

```python
y.shape
plt.figure()
plt.plot(y[:,0])
```

```python
spectra = np.vstack((s1[:,1],s2[:,1]))
spectra = np.transpose(spectra)
spectra.shape
```

```python
reg = linear_model.Lasso(positive=True,alpha=0.08)
reg.fit(spectra, y)
reg.coef_.shape
```

```python
x1 = reg.coef_[:,0]
x2 = reg.coef_[:,1]
x1 = np.reshape(x1,[120,80])
x2 = np.reshape(x2,[120,80])
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(x1)
plt.subplot(122)
plt.imshow(x2)
```
